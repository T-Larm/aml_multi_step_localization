```python
import pickle
import pandas as pd
import numpy as np

with open('./predictions/person_ep030_all.pkl', 'rb') as f:
    results = pickle.load(f)

df = pd.DataFrame(results)
print(f"总预测数: {len(df)}")
print(f"视频数: {df['video-id'].nunique()}")
print(f"\nscore 分布:")
print(df['score'].describe())

print(f"\n不同threshold下平均每视频步数:")
for thresh in [0.05, 0.1, 0.2, 0.25,.26,0.27,0.28,0.29,0.3,0.4,0.5]:
    kept = df[df['score'] >= thresh]
    avg_steps = kept.groupby('video-id').size().mean()
    print(f"threshold={thresh}: 平均 {avg_steps:.1f} 步/视频")
```

```python
import pickle
import pandas as pd
import numpy as np

with open('./predictions/recordings_ep030_all.pkl', 'rb') as f:
    results = pickle.load(f)

df = pd.DataFrame(results)
print(f"总预测数: {len(df)}")
print(f"视频数: {df['video-id'].nunique()}")
print(f"\nscore 分布:")
print(df['score'].describe())

print(f"\n不同threshold下平均每视频步数:")
for thresh in [0.05, 0.1, 0.2, 0.25,.26,0.27,0.28,0.29,0.3,0.4,0.5]:
    kept = df[df['score'] >= thresh]
    avg_steps = kept.groupby('video-id').size().mean()
    print(f"threshold={thresh}: 平均 {avg_steps:.1f} 步/视频")
```

```python
import numpy as np

data = np.load('./step_embeddings/person_ep030.npz')
counts = np.array([data[vid].shape[0] for vid in data.files])

print(f'总视频数: {len(counts)}')
print(f'Min=1 的视频数: {(counts == 1).sum()}')
print(f'步数<5 的视频数: {(counts < 5).sum()}')
print(f'步数=50 的视频数: {(counts == 50).sum()}')
print(f'平均步数: {counts.mean():.1f}')
print(f'最小步数: {counts.min()}')
print(f'最大步数: {counts.max()}')
```
